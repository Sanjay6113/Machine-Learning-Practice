```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import ipywidgets as widgets

# Data storage
x_data = []
y_data = []

# Input widgets
x_input = widgets.FloatSlider(value=0.5, min=0, max=1, step=0.01, description='X:')
y_input = widgets.FloatSlider(value=0.5, min=0, max=1, step=0.01, description='Y:')
add_button = widgets.Button(description="Add Point")
clear_button = widgets.Button(description="Clear All")

output = widgets.Output()

def plot():
    with output:
        clear_output(wait=True)
        fig, ax = plt.subplots(figsize=(6,6))
        ax.set_xlim(0,1)
        ax.set_ylim(0,1)
        ax.set_title('Linear Regression Demo')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')

        # Plot points
        ax.scatter(x_data, y_data, color='blue')

        # Draw regression line if enough points
        if len(x_data) >= 2:
            A = np.vstack([x_data, np.ones(len(x_data))]).T
            m, c = np.linalg.lstsq(A, y_data, rcond=None)[0]
            x_line = np.linspace(0,1,100)
            y_line = m * x_line + c
            ax.plot(x_line, y_line, color='red', linewidth=2)

        plt.show()

def add_point(b):
    x_data.append(x_input.value)
    y_data.append(y_input.value)
    plot()

def clear_all(b):
    x_data.clear()
    y_data.clear()
    plot()

add_button.on_click(add_point)
clear_button.on_click(clear_all)

# Display UI
ui = widgets.VBox([x_input, y_input, widgets.HBox([add_button, clear_button]), output])
display(ui)

# Initial plot
plot()
```
