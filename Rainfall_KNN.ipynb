```python
# REGRESSION MODEL
import pandas as pd
df = pd.read_csv('/content/chennai_rainfall_2019_2023_mm.csv')
df = df.dropna()
from sklearn.preprocessing import MinMaxScaler
regr = MinMaxScaler()
df[['Humidity (%)','Temperature (°C)','Wind Speed (km/h)']] = regr.fit_transform(df[['Humidity (%)','Temperature (°C)','Wind Speed (km/h)']])
X = df[['Humidity (%)','Temperature (°C)','Wind Speed (km/h)']]
Y = df['Rainfall (mm)']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=2)
neigh.fit(X_train, y_train)
y_pred = neigh.predict(X_test)

from sklearn.metrics import accuracy_score, r2_score, mean_squared_error
import numpy as np
r2 = r2_score(Y_test, y_pred)
rmse = np.sqrt(mean_squared_error(Y_test, y_pred))
print(f'R² Score: {r2:.4f}')
print(f'RMSE: {rmse:.4f}')
```
